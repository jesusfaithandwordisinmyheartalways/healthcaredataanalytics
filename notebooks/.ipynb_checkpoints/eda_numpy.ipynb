```python
from openpyxl import load_workbook
import numpy as np

# Load cleaned Excel file
wb = load_workbook('../data/healthcare_data.xlsx')
sheet = wb.active

# Convert sheet to NumPy array (skip header)
data = np.array([[cell.value for cell in row] for row in sheet.iter_rows(min_row=2)])

# Billing amounts (column 6)
billing = data[:, 6].astype(float)

# Basic stats
print("Total Billing:", np.sum(billing))
print("Average Billing:", np.mean(billing))
print("Max Billing:", np.max(billing))

# Visits per Department (column 5)
departments = data[:, 5]
unique_departments, counts = np.unique(departments, return_counts=True)
for dept, count in zip(unique_departments, counts):
    print(f"{dept}: {count} visits")
```
